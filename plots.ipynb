Line Graph

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = {'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
        'Value': [10, 15, 20, 25, 30, 28, 32, 34, 40, 42]}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Value'], marker='o')
plt.title('Line Chart Example')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

This code first imports pandas for handling data and **matplotlib** for creating visualizations. It then creates sample data consisting of a sequence of dates and corresponding numerical values. The data is converted into a pandas DataFrame, which organizes it into a tabular format. Next, a plotting figure is created with a specified size for better visibility. The code plots a line chart using dates on the x-axis and values on the y-axis, with markers showing each data point. Finally, it adds a title, axis labels, grid lines, and displays the chart.


Bar chart


```python
import matplotlib.pyplot as plt

# Sample Data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

This code uses matplotlib to create a bar chart for visualizing categorical data. It first defines a list of categories and a corresponding list of numerical values. A figure is created with a specified size to make the chart clear and readable. The plt.bar() function draws vertical bars for each category, with their heights representing the values. The bars are colored sky blue for better visual appeal. Finally, a title and axis labels are added, and the chart is displayed on the screen.


Scatter Plot


```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()

```

This code uses the Matplotlib library to create a simple scatter plot. First, it imports matplotlib.pyplot for plotting graphs. Then, two lists x and y are defined to store the data points. The plt.figure() function sets the size of the plot window. The plt.scatter() function plots the points as red dots on the graph. The title and labels for the X and Y axes are added using plt.title(), plt.xlabel(), and plt.ylabel(). Finally, plt.show() displays the scatter plot on the screen.


Bubble Chart

```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
sizes = [100, 200, 300, 400, 500]

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, alpha=0.5, color='green')
plt.title('Bubble Chart Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

This code uses the Matplotlib library to create a bubble chart. First, matplotlib.pyplot is imported to enable plotting. The lists x and y store the coordinates of the data points, while the sizes list controls the size of each bubble. The plt.figure() function sets the size of the chart. The plt.scatter() function plots the points as green bubbles, where s=sizes defines bubble sizes and alpha=0.5 makes them semi-transparent. The title and axis labels are added for clarity, and finally plt.show() displays the bubble chart on the screen.

Area Plot and Stacked Plot


```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'A': [10, 15, 20, 25, 30, 28, 32, 34, 40, 42],
    'B': [5, 7, 8, 6, 7, 5, 8, 9, 10, 11]
})

# Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(data['Date'], data['A'], color="skyblue", alpha=0.4, label='A')
plt.fill_between(data['Date'], data['B'], color="salmon", alpha=0.4, label='B')
plt.title('Area Plot Example')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

This code creates an area plot using Matplotlib and Pandas. First, the required libraries are imported, and a Pandas DataFrame is created containing dates and two sets of values, A and B. The pd.date_range() function generates a sequence of dates for the X-axis. The plt.figure()function sets the size of the plot. The plt.fill_between()function is used twice to fill the area under the curves for values A and B with different colors and transparency. The title, axis labels, and legend are added for better understanding, and finally plt.show() displays the area plot.


Pie Chart

```python
import matplotlib.pyplot as plt

# Sample Data
labels = ['Category A', 'Category B', 'Category C']
sizes = [40, 30, 30]

# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Pie Chart Example')
plt.show()
```

This code uses the Matplotlib library to create a pie chart. First, matplotlib.pyplot is imported for plotting. The labels list defines the names of each category, and the sizes list represents their corresponding values. The plt.figure() function sets the size of the chart. The plt.pie() function draws the pie chart, where labels name each slice, autopct displays the percentage on each slice, and colors assigns different colors to them. Finally, a title is added and plt.show() displays the pie chart on the screen.


Table Chart


```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Value': [40, 30, 30]
})

# Plot
fig, ax = plt.subplots(figsize=(6, 2))
ax.axis('tight')
ax.axis('off')
ax.table(cellText=data.values, colLabels=data.columns, cellLoc='center', loc='center')
plt.title('Table Chart Example')
plt.show()

```

This Python code creates and displays a **table-style chart** using pandas and **matplotlib**. First, it imports matplotlib.pyplot for plotting and pandas for handling tabular data. Then, a sample DataFrame named data is created with two columns: Category (A, B, C) and their corresponding Value (40, 30, 30). After that, a figure and axes are created with a small size using plt.subplots(). The axes are turned off using ax.axis('off') so that no graph lines or ticks appear. The ax.table() function is used to display the DataFrame values as a table in the center of the figure, with column labels and centered text inside each cell. Finally, a title is added to the table using plt.title(), and plt.show() displays the table on the screen.


Polar Chart


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
labels=np.array(['A', 'B', 'C', 'D'])
values=np.array([4, 3, 2, 5])
angles=np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()

# Complete the loop
values=np.concatenate((values,[values[0]]))
angles+=angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='cyan', alpha=0.5)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Polar Chart Example')
plt.show()
```

This code creates a polar (radar/spider) chart using Matplotlib. First, it imports matplotlib.pyplot for plotting and numpy for numerical operations. The labels array stores the category names (A, B, C, D) and the values array stores their corresponding numeric values. The angles are calculated using np.linspace to divide a full circle (0 to 2 radians) into equal partsone for each label. To make the radar chart a **closed shape**, the first value is appended again at the end of the values array, and the first angle is added to the end of the angles list. Then a polar subplot is created using subplot_kw=dict(polar=True). The ax.fill() function draws and fills the polygon formed by the angles and values with a semi-transparent cyan color. The y-axis tick labels are hidden for a cleaner look, the x-axis ticks are set to the angles of the labels, and the category names are displayed around the circle. Finally, a title is added and plt.show() displays the polar chart.


Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
data = np.random.randn(1000)

# Plot
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

This code generates and displays a histogram to visualize the distribution of data using Matplotlib. First, it imports matplotlib.pyplot for plotting and numpy for numerical operations. The data variable is created using np.random.randn(1000), which generates 1000 random values following a normal (Gaussian) distribution with a mean of 0 and a standard deviation of 1. A figure is then created with a specified size to make the plot clear and readable. The plt.hist() function draws the histogram by dividing the data into 30 bins, where each bin represents a range of values, and the height of each bar shows how many data points fall within that range. The bars are colored purple with black edges to improve visibility. Finally, a title and axis labels are added to explain the chart, and plt.show()`displays the histogram on the screen.


Lollipop Chart

```python
import matplotlib.pyplot as plt

# Sample Data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Plot
plt.figure(figsize=(10, 6))
plt.stem(categories, values, basefmt=" ")
plt.title('Lollipop Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

This code creates a lollipop chart using Matplotlib to compare values across different categories. First, it imports matplotlib.pyplot for plotting. The categories list contains the category labels (A, B, C, D) and the values list holds their corresponding numerical values. A figure is created with a specified size to make the chart clear. The plt.stem() function is used to draw the lollipop chart, where vertical lines (stems) extend from a baseline up to each value and a marker is placed at the top, visually resembling lollipops. The basefmt option hides the baseline, and use_line_collection=True improves performance and rendering. Finally, a title and labels for both axes are added, and plt.show()displays the lollipop chart.
